## 数字图像基础

---

### 简单的图像形成模型

利用$f(x,y)$的二维函数来表示图像。$(x,y)$表示图像的空间坐标，$f$则是一个正标量值（其物理意义有图像源来决定）。

$$0<f(x,y)<+\infty$$

$(x_0, y_0)$处的强度（灰度）表示为：

$$
l = f(x_0,y_0)\\
L_{min} \leq l \leq L_{max}
$$

区间$[L_{min}, L_{max}]$称为_灰度级_（或_强度级_）。实际情况下令该区间为$[0, L-1]$, 其中$l=0$为黑色，$l=L-1$在灰度级中为白色。所有中间值时从黑色到白色之间的变化的灰度色调。

### 数字图像的表示

$M\times N$的实数字阵列：

$$f(x,y) = \left[\begin{array}{}
f(0,0) & f(0,1)& \cdots & f(0, N-1)\\
f(1,0) & f(1,1)& \cdots & f(1, N-1)\\
\cdots & \cdots &\cdots & \cdots\\
f(M-1, 0) & f(M-1, 1) & \cdots & f(M-1, N-1)
\end{array}\right]$$

等效的表达了一副数字图像。数字整列中的每一个元素称为图像单元、图像元素或像素。

利用传统的矩阵来表示图像和其像素更为方便：

$$
A = \left[\begin{array}{}
a_{0,0} & a_{0,1}& \cdots & a_{0, N-1}\\
a_{1,0} & a_{1,1}& \cdots & a_{1, N-1}\\
\cdots & \cdots &\cdots & \cdots\\
a_{M-1, 0} & a_{M-1, 1} & \cdots & a_{M-1, N-1}
\end{array}\right]
$$

用矩阵来表示图像需要对图像进行量化，要对$M,N$以及灰度级数$L$做出判定。$M,N$只要求为正整数，出于存储和量化硬件的考虑，灰度级通常取为2的整数次幂，即：

$$
L = 2^k
$$

**对比度：**一图像中的最高和最低灰度间的灰度差定义为对比度。

**图像的存储比特数$b$：**

$$
b = M \times N \times k
$$

当$M=N$时：

$$
b = N^2k
$$

### 图像的取样和量化

#### 空间和灰度分辨率

**空间分辨率：**单位距离的线对数或单位距离的点数（像素数）。直观描述为图像中可分辨的最小细节的度量。该度量通常使用**_每英寸电点数(dpi)_**来表示。

常用空间分辨率：

|用途|空间分辨率（单位：dpi）|
|---|---|
|报纸|75|
|杂志|133|
|光鲜的小册子|175|
|书|2044|

**灰度分辨率：**在灰度级中可分辨的最小变化，用于量化灰度的比特数。

#### 图像的内插

**内插**是诸如放大、收缩、旋转和几何校正等任务中广泛应用的工具。
